## Importing all the required libraries

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y = make_regression(n_samples=100,
                n_features=1,
                n_informative=1,
                n_targets=1,
                noise=20,
                random_state=101)
```

```python
plt.scatter(X,y)
plt.show()
```

## Lets fit the plane Linear Regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X,y)

print(lr.coef_)
print(lr.intercept_)
```

## Now lets fit it in our Ridge function

```python
from sklearn.linear_model import Ridge
```

```python
lrr = Ridge(alpha=10)
lrr.fit(X,y)

print(lrr.coef_)
print(lrr.intercept_)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X), color='red', label='alpha = 0')
plt.plot(X,lrr.predict(X), color='green', label='alphs = 10')
```
